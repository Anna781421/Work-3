```
from ludwig.api import LudwigModel
import pandas as pd
```

```
# Таблица с данными
ocean = pd.read_csv("test-results/data-final.csv",
                    sep='\t')
ocean.head()
```

```
# Пусть будет предсказание страны по первым 10 вопросам опросника
model_dict = {"input_features":
              [{'name': "EXT1", "type": 'numerical'},
               {'name': "EXT2", "type": 'numerical'},
               {'name': "EXT3", "type": "numerical"},
               {'name': 'EXT4', "type": 'numerical'},
               {'name': 'EXT5', "type": 'numerical'},
               {'name': 'EXT6', "type": 'numerical'},
               {'name': 'EXT7', "type": 'numerical'},
               {'name': 'EXT8', "type": 'numerical'},
               {'name': 'EXT9', "type": 'numerical'},
               {'name': 'EXT10', "type": 'numerical'}],
              "output_features":
              [{"name": "country", "type": 'category'}]
              }

# Создаём модель
model = LudwigModel(model_dict)

# Обучаем (Много данных, это долго)
train_stats = model.train(ocean)
```
